# Loading TNG50 MW/M31 Analogs cutouts

This is a short tutorial to get started with the cutouts from the MW/M31 analogs ([Pillepich+2023](https://arxiv.org/abs/2303.16217)), and walks you through making a first plot with the outputs.
The documentation is available at https://www.tng-project.org/data/milkyway+andromeda/. 

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import matplotlib as mpl
import illustris_python as il
```

# Specify galaxy and snapshot

Here loading the final snapshot at $z=0$ (99) for the subhalo 342447.

```python
snapshot = 99
subhaloID = 342447
```

First, let's get all the general properties of this subhalo (i.e. galaxy) from the catalog:

```python
subhalo = il.groupcat.loadSingle('sims.TNG/TNG50-1/output/', snapNum=snapshot, subhaloID=subhaloID)
```

```python
log_mstar = np.log10(subhalo['SubhaloMassInRadType'][4] * 1e10 / 0.6774)
print(log_mstar)
```

# Local MW/M31 cutout data

Since we are working on the TNG JupyterLab, there is no need to download data, we can access it directly. We specify the path, and open the file.

```python
path = 'sims.TNG/TNG50-1/postprocessing/MWM31s/'
filepath = path + 'cutouts/snap_%03d/%d.hdf5' % (snapshot, subhaloID)
```

```python
cutout = h5py.File(filepath,'r')
```

Then explore the datasets which the file contains:

```python
fields = cutout.keys()
    
print(fields)
```

OK, we have the several types of particles: 0 for gas, 1 for DM and 4 for stars. Others are: tracer particles and black holes.

```python
stars = cutout['PartType4']
```

```python
stars.keys()
```

# A basic analysis/visualization

Make a histogram of the star particles in box comoving coordinates

```python
nbins = 400

plt.figure()
xb = np.linspace(subhalo['SubhaloCM'][0] - 40, subhalo['SubhaloCM'][0] + 40, nbins)
yb = np.linspace(subhalo['SubhaloCM'][1] - 40, subhalo['SubhaloCM'][1] + 40, nbins)
plt.hist2d(stars['Coordinates'][:,0], stars['Coordinates'][:,1], 
           bins=(xb,yb), norm=mpl.colors.LogNorm())
plt.gca().set_aspect('equal')
plt.xlabel(r'x [ckpc/h]')
plt.ylabel(r'y [ckpc/h]');
```

Now we can even make face-on and edge-on views, without calculating anything, thanks to the rotated coordinates (available in the MW/M31 cutouts):

```python
plt.figure()
xb = np.linspace(-40, +40, nbins)
yb = np.linspace(-40, +40, nbins)
plt.hist2d(stars['RotatedCoordinates'][:,0], stars['RotatedCoordinates'][:,1], 
           bins=(xb,yb), norm=mpl.colors.LogNorm())
plt.gca().set_aspect('equal')
plt.xlabel(r'x [ckpc/h]')
plt.ylabel(r'y [ckpc/h]');
```

```python
plt.figure()
xb = np.linspace(-40, +40, nbins)
yb = np.linspace(-40, +40, nbins)
plt.hist2d(stars['RotatedCoordinates'][:,0], stars['RotatedCoordinates'][:,2], 
           bins=(xb,yb), norm=mpl.colors.LogNorm())
plt.gca().set_aspect('equal')
plt.xlabel(r'x [ckpc/h]')
plt.ylabel(r'y [ckpc/h]');
```

And the dark matter particle (i.e. mass) distribution (i.e. 2D histogram) over the full 400 kpc cutout (in the galaxy face-on rotation):

```python
dm = cutout['PartType1']

plt.figure()
xb = np.linspace(-400, +400, nbins)
yb = np.linspace(-400, +400, nbins)
plt.hist2d(dm['RotatedCoordinates'][:,0], dm['RotatedCoordinates'][:,1], 
           bins=(xb,yb), norm=mpl.colors.LogNorm())
plt.gca().set_aspect('equal')
plt.xlabel(r'x [ckpc]')
plt.ylabel(r'y [ckpc]');
```

```python
# good to close the file when we're done
cutout.close()
```
